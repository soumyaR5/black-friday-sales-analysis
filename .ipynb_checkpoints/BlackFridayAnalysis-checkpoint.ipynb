```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Black Friday Data 
df = pd.read_csv('BlackFriday.csv')
df.head()
```

```python
#Data Cleaning
```

```python
#handling missing values
print(df.isnull().sum())
```

```python
#checking which column needs data type fixing
df.dtypes
```

```python
df['Product_Category_2'] = df['Product_Category_2'].fillna(df['Product_Category_2'].mean())
df['Product_Category_3'] = df['Product_Category_3'].fillna(df['Product_Category_3'].mean())
#removing duplicate values
df = df.drop_duplicates()
#converting age to numeric bins
df['Age_bin'] = df['Age'].map({
    '0-17': 1,
    '18-25': 2,
    '26-35': 3,
    '36-45': 4,
    '46-50': 5,
    '51-55': 6,
    '55+': 7
})
df['Product_Category_2'] = df['Product_Category_2'].astype(int)
df['Product_Category_3'] = df['Product_Category_3'].astype(int)
df['Purchase'] = df['Purchase'].astype(float)
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].str.extract(r'(\d+)').astype(int)
df['Gender'] = df['Gender'].astype('category')
df['City_Category'] = df['City_Category'].astype('category')
```

```python
#After data cleaning
df.head()
```

```python
df.dtypes
```

```python
#Performing exploratory data analysis
#Customer demographics
#which age group spends the most?
sns.barplot(x = df['Age'], y = df['Purchase'])
plt.xlabel('Age Group')
plt.ylabel('Purchase')
plt.title('Purchase per Age')
plt.show()
```

```python
#According to the above chart,
#Age group between 51-55 spends the most
```

```python
#Do male or female purchase more?
sns.countplot(x='Gender',data = df)
plt.title('Total purchase by gender')
plt.show()
```

```python
#which city category has highest spending
sns.countplot(x = 'City_Category', data = df)
plt.title('Total Purchase by City Category')
plt.show()
```

```python
#Product Insights
#Top 10 best selling product categories

num_cols = pd.concat([
    df['Product_Category_1'],
    df['Product_Category_2'],
    df['Product_Category_3']
])
top10 = num_cols.value_counts().head(10)
sns.barplot(x = top10.values, y = top10.index.astype(str))
plt.xlabel('Purchase')
plt.ylabel('Product Category')
plt.title('Top 10 best selling product categories')
plt.show()
```

```python
#Which category has highest revenue
df_cat1 = df[['Product_Category_1','Purchase']].rename(
    columns = {'Product_Category_1': 'Category'}
)
df_cat2 = df[['Product_Category_2','Purchase']].rename(
    columns = {'Product_Category_2':'Category'}
)
df_cat3 = df[['Product_Category_3','Purchase']].rename(
    columns = {'Product_Category_3':'Category'}
)
all_cat = pd.concat([df_cat1,df_cat2,df_cat3],ignore_index = True)
all_cat = all_cat.dropna(subset=['Category'])

revenue = all_cat.groupby('Category')['Purchase'].sum().sort_values(ascending = False)

sns.barplot(x=revenue.values, y=revenue.index.astype(str))
plt.title('Product Categories by Revenue')
plt.xlabel('Revenue')
plt.ylabel('Category')
plt.show()
```

```python
#Behavioral Pattern
#Relationship between years in city vs purchase amount
sns.boxplot(x = 'Stay_In_Current_City_Years',y = 'Purchase',data = df)
plt.show()
```

```python
#Average purchase per user vs per product
avg_user = df.groupby('User_ID')['Purchase'].mean()
avg_product = df.groupby('Product_ID')['Purchase'].mean()
data = [avg_user.mean(),avg_product.mean()]
label = ['Avg user per purchase','Avg product per purchase']

sns.barplot(x = label, y = data)
plt.xlabel('Average user & product')
plt.ylabel('Purchase')
plt.title('Average Purchase per user vs per product')
plt.show()
```

```python
#Distribution Analysis
#Purchase amount distribution
sns.histplot(df['Purchase'],bins=10,kde = True)
plt.xlabel('Purchase Amount')
plt.title('Purchase Amount Distribution')
plt.ylabel('Frequency')
plt.show()
```
